```python
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer

# Read the data from file using read_csv
data = pd.read_csv('stomp.csv', sep=',', header=0)

# Print the first 5 rows of the dataframe
print(data.head())
```

```python
# COnsider only first letter in each row
for i in range(data.shape[1]):
    data.iloc[:,i] = data.iloc[:,i].str[0]
```

```python
data
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler, LabelEncoder
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

```

```python
data_scaled
```

```python

from sklearn.decomposition import PCA
from numpy import sum

n= data_scaled.shape[1]
n = 6

pca = PCA(n_components=n)
pca.fit(data_scaled)
data_pca = pca.transform(data_scaled)

# print("pca.components")
# print(pca.components_.T.round(2))
# print(variance_explained.round(3))
variance_explained = pca.explained_variance_

fa = FactorAnalyzer(n_factors=n,method='principal', rotation='varimax')
fa.fit_transform(data_scaled)
var = fa.get_factor_variance()


loading = pd.DataFrame(fa.loadings_,columns=[f"Component{i+1}" for i in range(n)])
loading.insert(0,"Genre",data.columns)
print( loading.round(2))

# var_to = np.asarray(var[0])
# var_to = np.var(fa.loadings_, axis=0)
print("\n variance of varimax")
print(var[0])

# print("\n varimax matrix")
# print(fa.loadings_.round(2))
# print(fa.loadings_.shape)


# print(loadings.shape)
# print(loadings.round(2).T)


```
