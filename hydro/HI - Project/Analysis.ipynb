```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading data
data = pd.read_csv('./HI-dataset.csv')
```

```python
# data cleaning
data.columns = data.columns.str.strip()
```

```python
#data conversion
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
```

1. Simple Ananlysis visulaisation using graphs


```python
data.sort_values(by='Date', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Manjeera water drawn from HMWSSB'].values, label='Manjeera Water Supply', color='skyblue')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Manjeera Water Supply Trend')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlim(data['Date'].min(), data['Date'].max())
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Water drawn from borewells'].values, label='Ground level', color='lightgreen')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Ground level Trend')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlim(data['Date'].min(), data['Date'].max())
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Total water consumption'].values, label='Total Water Consumption', color='red')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Total Water Consumption Trend')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlim(data['Date'].min(), data['Date'].max())
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Manjeera water drawn from HMWSSB'].values, label='Manjeera Water Supply', color='skyblue')
plt.plot(data['Date'], data['Water drawn from borewells'].values, label='Groundwater Level', color='lightgreen')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Manjeera Water Supply and Groundwater Level Trends')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlim(data['Date'].min(), data['Date'].max())

plt.show()
```

2. Bar plots

```python
plt.figure(figsize=(12, 6))

# Manjeera Water Supply bar plot
plt.subplot(1, 3, 1)
plt.bar(data['Date'], data['Manjeera water drawn from HMWSSB'].values, color='blue')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Manjeera Water Supply')

# Groundwater Level bar plot
plt.subplot(1, 3, 2)
plt.bar(data['Date'], data['Water drawn from borewells'].values, color='green')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Groundwater Level')

# Total Water Consumption bar plot
plt.subplot(1, 3, 3)
plt.bar(data['Date'], data['Total water consumption'].values, color='red')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Total Water Consumption')

plt.tight_layout()

# Show plot
plt.show()
```

3. Histograms

```python

# Plotting histograms
plt.figure(figsize=(12, 6))

# Manjeera Water Supply histogram
plt.subplot(1, 3, 1)
plt.hist(data['Manjeera water drawn from HMWSSB'].values, color='blue', bins=20)
plt.xlabel('Volume (units)')
plt.ylabel('Frequency')
plt.title('Manjeera Water Supply Distribution')

# Groundwater Level histogram
plt.subplot(1, 3, 2)
plt.hist(data['Water drawn from borewells'].values, color='green', bins=20)
plt.xlabel('Volume (units)')
plt.ylabel('Frequency')
plt.title('Groundwater Level Distribution')

# Total Water Consumption histogram
plt.subplot(1, 3, 3)
plt.hist(data['Total water consumption'].values, color='red', bins=20)
plt.xlabel('Volume (units)')
plt.ylabel('Frequency')
plt.title('Total Water Consumption Distribution')

plt.tight_layout()

# Show plot
plt.show()
```

4. Stacked Bar plot

```python
plt.figure(figsize=(10, 6))

# Calculate the total volume for each date
total_manjeera = data['Manjeera water drawn from HMWSSB'].values
total_groundwater = data['Water drawn from borewells'].values
dates = data['Date']

# Plotting stacked bar plot with thicker bars
plt.bar(dates, total_manjeera, label='Manjeera Water Supply', color='skyblue', width=3)
plt.bar(dates, total_groundwater, bottom=total_manjeera, label='Groundwater Level', color='pink', width=3)

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Manjeera Water Supply and Groundwater Level')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

5. box plot


```python
plt.figure(figsize=(10, 6))
plt.boxplot([data['Manjeera water drawn from HMWSSB'].values, data['Water drawn from borewells'].values])
plt.xticks([1, 2], ['Manjeera Water Supply', 'Groundwater Level'])
plt.ylabel('Volume (units)')
plt.title('Distribution of Manjeera Water Supply and Groundwater Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

6. Voilin Plot

```python
plt.figure(figsize=(10, 6))
plt.violinplot([data['Manjeera water drawn from HMWSSB'].values, data['Water drawn from borewells'].values], showmedians=True)
plt.xticks([1, 2], ['Manjeera Water Supply', 'Groundwater Level'])
plt.ylabel('Volume (units)')
plt.title('Distribution of Manjeera Water Supply and Groundwater Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

7. HeatMap - correlation

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(data[['Manjeera water drawn from HMWSSB', 'Water drawn from borewells', 'Total water consumption']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

8. Area Map

```python
plt.figure(figsize=(10, 6))
plt.fill_between(data['Date'], data['Manjeera water drawn from HMWSSB'].values, color='skyblue', alpha=0.5, label='Manjeera Water Supply')
plt.fill_between(data['Date'], data['Water drawn from borewells'].values, color='green', alpha=0.5, label='Groundwater Level')
plt.xlabel('Date')
plt.ylabel('Volume (units)')
plt.title('Manjeera Water Supply and Groundwater Level')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

9. Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Manjeera water drawn from HMWSSB'], data['Water drawn from borewells'], color='purple')
plt.xlabel('Manjeera Water Supply')
plt.ylabel('Groundwater Level')
plt.title('Scatter Plot of Manjeera Water Supply vs Groundwater Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```
