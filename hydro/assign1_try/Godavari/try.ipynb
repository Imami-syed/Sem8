```python
import os
import pandas as pd
import csv

# Directory containing CSV files
directory = "./godavari_basins"

# List to store DataFrames
dataframes = []
num_files=0
# Iterate over files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        num_files+=1
        print(f"{filename}")
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        # Replace zeros with a very small value
        # df['ACTUAL (mm)'] = df['ACTUAL (mm)'].replace(0, 1e-15)
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes)

# Display combined DataFrame
print(f"Number of files : ", num_files)
# print(73*num_files)
print(combined_df)

```

```python
import os
import pandas as pd
import tensorflow as tf

# Directory containing CSV files
directory = "./godavari_basins"

# List to store DataFrames
dataframes = []

# Iterate over files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        # Replace zeros with a very small value
        # df['ACTUAL (mm)'] = df['ACTUAL (mm)'].replace(0, 1e-15)
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes)

# Convert DataFrame to TensorFlow tensor
tensor_data = tf.constant(combined_df.values)

# Print TensorFlow tensor
print(tensor_data)
```
