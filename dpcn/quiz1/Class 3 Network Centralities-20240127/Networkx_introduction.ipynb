```python
import numpy as np
import networkx as nx
```

```python
N=8
G=nx.Graph()
```

```python
a=['D','I','S','C','A','DU','N','H']
```

```python
for i in range(N):
  G.add_node(a[i])
```

```python
nx.draw(G, with_labels=True)
```

```python
G.add_edges_from([(a[0],a[4])])
```

```python
nx.draw(G, with_labels=True)
```

```python
G.add_edges_from([(a[0], a[1]),(a[1], a[2]),(a[2], a[3]),(a[2], a[5]),
                  (a[3], a[5]),(a[5], a[6]),(a[6], a[4]),(a[6], a[7])])
```

```python
nx.draw(G, with_labels=True)
```

```python
A=nx.to_numpy_array(G)
A=np.array(A)
```

```python
A
```

```python
(A==A.T).all()
```

```python
degree=np.sum(A, axis=1)
L=degree*np.eye(N)-A
```

```python
L
```

```python
p,q=np.linalg.eig(L)
```

```python
p=np.sort(p)
print(p)
```

```python
path=nx.single_source_shortest_path(G,'H')
print(path)
```

```python
average_path=nx.average_shortest_path_length(G)
print(average_path)
```

```python
degree=np.sum(A, axis=0)
```
